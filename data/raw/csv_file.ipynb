```python
import json
import pandas as pd

# Đọc file JSON
with open(r"D:\steam-market-analytics-platform\data\raw\steamspy_page_0.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# Chuyển dictionary thành list record
records = list(data.values())

# Tạo DataFrame
df = pd.DataFrame(records)

# Xuất CSV
df.to_csv("steam_games.csv", index=False, encoding="utf-8-sig")

print("Đã xuất steam_games.csv")
```
